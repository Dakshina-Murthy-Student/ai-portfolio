```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic') # tiny, friendly dataset
print("Rows, cols:", df.shape)
df.head()
```

```python
display(df.head()) # 1st rows
print("\nColumn names:", list(df.columns))
display(df.describe(include= 'all')) # numeric + non-numeric summary
```

```python
# Cell 3: small transform and save a CSV locally in Colab (you can download it)
df_clean = df[['survived','pclass','sex','age','fare']].dropna()
print("After selecting columns & dropping NA:", df_clean.shape)
df_clean.to_csv('titanic_clean.csv', index=False)
print("Saved titanic_clean.csv — you can download it via Colab Files panel.")
```
